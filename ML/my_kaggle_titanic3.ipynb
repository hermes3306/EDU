```python
# Age 값이 Null 값이면 전체 값의 평균 값을 대입한다.
import math 

n_of_n = train['Age'].isnull().count()   # all
n_of_t = train['Age'].count()            # not null count

sum = 0
arr = []
for s in train['Age'].values:
    if(math.isnan(s)) :
        arr.append(0)
    else:
        arr.append(s)
        sum = sum + s
        
avg = sum / n_of_t
avg = round(avg,0)
print(sum)
print(avg)

arr2 = []
for i in arr:
    if(i==0): 
        arr2.append(avg)
    else:
        arr2.append(i)
        
train['Age'] = arr2
```
