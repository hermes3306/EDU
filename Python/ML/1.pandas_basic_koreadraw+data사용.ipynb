```python
def get_max_pop_df(sido_name):
    max_pop = data.loc[data['광역시도'] == sido_name,'population'].max()
    return data.loc[data['population'] == max_pop,['광역시도','행정구역','population']]
```

```python
get_max_pop_df('경상북도')
```

```python
#새로운 DataFrame을 생성
max_pop_df = \
pd.DataFrame(columns=['광역시도','행정구역','population'])

for sido_name in data['광역시도'].unique():
    max_data = get_max_pop_df(sido_name)
    max_pop_df = max_pop_df.append(max_data)
    
max_pop_df
```

```python
# max_pop_df 의 index값을 reset
max_pop_df = max_pop_df.reset_index()
#index 컬럼을 삭제
del max_pop_df['index']
max_pop_df
```
